```python
# CODE READ DATA

import os

def read_data(path):
    X = []
    for file in os.listdir(path):
        image = imread(path + '/' + file)
        smallimage = cv2.resize(image, (224,224))
        print(path + '/' + file)

        X.append(smallimage)
    return np.array(X)
```

```python
X_test = read_data('img')

#Preprocesar las imagenes tal y como entran en el model
X_test = preprocess_input(X_test)
print(X_test.shape)
```

```python
# CODE preds

preds = base_model.predict(X_test)
```

## VGG16
En este caso vamos a importar la red VGG16, que utilizaremos como red preentrenada y completaremos con una fully connected layer. 

```python
import pandas as pd
from sklearn.model_selection import train_test_split

IM_SIZE=32

TRAIN_PATH = 'DATASET/TRAIN'
filenames = os.listdir(TRAIN_PATH)
categories = []
for filename in filenames:
    category = filename.split('.')[0]
    categories.append(category)
    
df = pd.DataFrame({
    'filenames': filenames,
    'category': categories
})

train_df, validate_df = train_test_split(df,
                                         test_size=0.20,
                                         random_state=42)

train_df = train_df.reset_index(drop=True)
validate_df = validate_df.reset_index(drop=True)
```

```python
from PIL import Image
import os

def convert_images_to_rgb(df, path):
    for fname in df['filenames']:
        full_path = os.path.join(path, fname)
        try:
            with Image.open(full_path) as img:
                if img.mode in ("P", "RGBA"):
                    img = img.convert("RGB")
                    img.save(full_path)
        except Exception as e:
            print(f"Error con {fname}: {e}")
```

```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split

IM_SIZE = 224
TRAIN_PATH = 'DATASET/TRAIN'






filenames = []
categories = []

for category in os.listdir(TRAIN_PATH):
    category_path = os.path.join(TRAIN_PATH, category)
    if os.path.isdir(category_path):
        for file in os.listdir(category_path):
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                filenames.append(f"{category}/{file}")
                categories.append(category)

df = pd.DataFrame({
    'filenames': filenames,
    'category': categories
})

train_df, validate_df = train_test_split(df, test_size=0.2, random_state=42)
train_df = train_df.reset_index(drop=True)
validate_df = validate_df.reset_index(drop=True)
print(df)
```

```python
convert_images_to_rgb(train_df, TRAIN_PATH)
convert_images_to_rgb(validate_df, TRAIN_PATH)
```

```python
train_df.head()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True


IM_SIZE = 224
BATCH_SIZE = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2  # usamos el 20% para validación
)

train_generator = train_datagen.flow_from_directory(
    directory='DATASET/TRAIN',
    target_size=(IM_SIZE, IM_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training'  # conjunto de entrenamiento
)

validation_generator = train_datagen.flow_from_directory(
    directory='DATASET/TRAIN',
    target_size=(IM_SIZE, IM_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'  # conjunto de validación
)

```

```python
from tensorflow.keras.applications.vgg16 import VGG16

# CODE model
base_model = VGG16(input_shape=(IM_SIZE,IM_SIZE,3),
                    include_top=False,
                    weights = 'imagenet'
                )
```

```python
len(base_model.layers)
```

```python
base_model.summary()
```

```python

##### FULLY CONNECTED LAYER #####
# Flatten the output layer to 1 dimension
x = layers.Flatten()(base_model.output)

# Add a fully connected layer with 512 hidden units and ReLU activation
x = layers.Dense(512, activation='relu')(x)

# Add a dropout rate of 0.5
x = layers.Dropout(0.5)(x)

# Add a final sigmoid layer for classification -> softmax porque hay más de 2 clases de salida
x = layers.Dense(5, activation='softmax')(x)

model = tf.keras.models.Model(base_model.input, x)

base_model.trainable = False
    
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy',metrics = ['acc'])
```

### Inception

```python
from PIL import Image
import os

folder = 'DATASET/TRAIN'

for subdir, _, files in os.walk(folder):
    for file in files:
        path = os.path.join(subdir, file)
        try:
            img = Image.open(path)
            img.verify()  # Verifica que no esté corrupta
        except:
            print(f"Eliminando imagen corrupta: {path}")
            os.remove(path)
```

```python
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras import layers,models

IM_SIZE = 224

base_model = InceptionV3(input_shape = (224, 224, 3),
                       include_top = False,
                       weights = None,
                       classifier_activation= "softmax")

for layer in base_model.layers:
    layer.trainable = False  # congelamos las capas del modelo base

x = layers.GlobalAveragePooling2D()(base_model.output)
x = layers.Dense(512, activation='relu')(x)
x = layers.Dropout(0.5)(x)
x = layers.Dense(5, activation='softmax')(x)



model = models.Model(inputs=base_model.input, outputs=x)
model.compile(optimizer = 'rmsprop',
             loss = 'categorical_crossentropy',
             metrics = ['accuracy'])
```

```python
model.fit(train_generator,
         validation_data = validation_generator,
         steps_per_epoch = 100,
         epochs = 5)
```

# Introducción a Machine Learning No Supervisado

Este notebook acompaña la sesión teórica sobre aprendizaje no supervisado y está diseñado para explorar conceptos básicos como clustering y reducción de dimensionalidad.

## Visualización del Dataset Iris

Primero, vamos a cargar y visualizar el dataset **Iris**, que contiene mediciones de largo y ancho de sépalo y pétalo. Aunque sabemos que tiene etiquetas, para este ejemplo ignoraremos esa información.

```python

import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Cargar dataset Iris
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)

# Visualizar datos
print("Primeras filas del dataset:")
display(data.head())

# Gráfico de dispersión (largo y ancho de pétalo)
plt.scatter(data.iloc[:, 2], data.iloc[:, 3], c='gray', alpha=0.7)
plt.title('Dispersión: Largo vs Ancho de Pétalo')
plt.xlabel('Largo de Pétalo')
plt.ylabel('Ancho de Pétalo')
plt.show()
```

## Reducción de Dimensionalidad con PCA

Utilizaremos PCA (Análisis de Componentes Principales) para reducir las dimensiones del dataset a solo dos componentes principales y visualizar el resultado.

```python

from sklearn.decomposition import PCA

# Aplicar PCA (2D)
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data)

# Visualizar resultados
plt.scatter(data_pca[:, 0], data_pca[:, 1], c='gray', alpha=0.7)
plt.title('Iris reducido a 2D con PCA')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.show()
```

## Actividad: Clustering manual

Dibujaremos puntos en un plano 2D y exploraremos cómo podríamos agruparlos manualmente antes de aplicar algoritmos de clustering.

```python

import numpy as np

# Crear puntos manualmente
np.random.seed(42)
points = np.random.rand(20, 2)

# Visualización inicial
plt.scatter(points[:, 0], points[:, 1], c='gray', s=50)
plt.title('Puntos sin etiquetar')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```
